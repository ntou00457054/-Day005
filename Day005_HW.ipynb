```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

import requests
import json

data = {"popular" : "true"}
r = requests.get("https://www.dcard.tw/_api/forums/pet/posts",params = data) #return  requests response module
response = r.text

data = json.loads(response)
print("æœƒå›å‚³ : "+ str(len(data)) + "ç­†è³‡æ–™") 
response
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

#print(type(data))
for d in data:
    print(d['title'])#æ¨™é¡Œ
    print(d['createdAt'])#è²¼æ–‡æ™‚é–“
    print(d['commentCount'])#ç•™è¨€äººæ•¸
    print(d['likeCount'])#æŒ‰è®šäººæ•¸
    print('\n')
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€


count_popular = 0
count_like = 0 
for d in  data:
    count_popular += d['commentCount']
    count_like += d['likeCount']
    
average_pop = count_popular / len(data) 
average_like = count_like / len(data) 
print("ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ : " + str(average_pop)  + "\n")
print("ç†±é–€æ–‡ç« æŒ‰è®šå¹³å‡äººæ•¸ : " + str(average_like)  + "\n")

data1 = {"popular" : "false"}
r1 = requests.get("https://www.dcard.tw/_api/forums/pet/posts")
data1 = json.loads(r1.text)

count_npopular = 0
count_nlike = 0 
for d in  data1:
    count_npopular += d['commentCount']
    count_nlike += d['likeCount']
average_npop = count_npopular / len(data1) 
average_nlike = count_nlike / len(data1) 
print("éç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸ : " + str(average_npop)  + "\n")
print("éç†±é–€æ–‡ç« æŒ‰è®šå¹³å‡äººæ•¸ : " + str(average_nlike)  + "\n")


```
