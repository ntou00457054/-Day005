## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json

data = {"popular" : "true"}
r = requests.get("https://www.dcard.tw/_api/forums/pet/posts",params = data) #return  requests response module
response = r.text

data = json.loads(response)
print("會回傳 : "+ str(len(data)) + "筆資料") 
response
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

#print(type(data))
for d in data:
    print(d['title'])#標題
    print(d['createdAt'])#貼文時間
    print(d['commentCount'])#留言人數
    print(d['likeCount'])#按讚人數
    print('\n')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」


count_popular = 0
count_like = 0 
for d in  data:
    count_popular += d['commentCount']
    count_like += d['likeCount']
    
average_pop = count_popular / len(data) 
average_like = count_like / len(data) 
print("熱門文章平均留言人數 : " + str(average_pop)  + "\n")
print("熱門文章按讚平均人數 : " + str(average_like)  + "\n")

data1 = {"popular" : "false"}
r1 = requests.get("https://www.dcard.tw/_api/forums/pet/posts")
data1 = json.loads(r1.text)

count_npopular = 0
count_nlike = 0 
for d in  data1:
    count_npopular += d['commentCount']
    count_nlike += d['likeCount']
average_npop = count_npopular / len(data1) 
average_nlike = count_nlike / len(data1) 
print("非熱門文章平均留言人數 : " + str(average_npop)  + "\n")
print("非熱門文章按讚平均人數 : " + str(average_nlike)  + "\n")


```
